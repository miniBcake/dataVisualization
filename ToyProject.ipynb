### (1) 강수량 대비 검색량을 조사

---

#### 1️⃣ **목적**
- 강수량 데이터와 부침개 검색량 데이터를 병합하여 두 데이터 간의 상관관계를 분석.
- 강수량 변화에 따라 부침개 검색량이 어떻게 변동하는지 확인.
- 2021년부터 2024년까지 연도별 데이터 비교를 통해 계절적 패턴 및 추이를 도출.

---

#### 2️⃣ **분석 개요**
- **강수량 데이터**는 기상청 데이터에서 추출된 날짜별 강수량(mm)을 기반으로 함.
- **검색량 데이터**는 네이버 검색 트렌드에서 추출된 날짜별 부침개 검색량을 기반으로 함.
- 날짜를 기준으로 데이터를 병합하여 강수량과 검색량이 같은 날 어떻게 변화하는지 확인.

---

#### 3️⃣ **분석 과정**
1. **데이터 변환**:
   - 강수량 데이터(`weatherArr`): 날짜(`'일시'`)를 `datetime` 형식으로 변환하여 연도별 분석 가능하도록 준비.
   - 검색량 데이터(`search_df`): 날짜(`'period'`)를 동일하게 변환하고, 검색량(`'searchVolume'`)을 기준으로 병합.

2. **날짜를 기준으로 병합**:
   - 날짜가 일치하는 강수량과 검색량 데이터만 남김으로써, 강수량이 검색량에 미친 영향을 분석 가능.

3. **연도별 데이터 분석**:
   - 연도별로 강수량과 검색량 데이터를 나누어 저장하여 연도 간 비교 및 추세 분석.

---

#### 4️⃣ **결과 기대**
- **강수량이 많은 날에 부침개 검색량이 증가하는 패턴을 관찰**.
- 계절적 요인(7월 장마철, 9월 추석 등)이 검색량 변화에 어떤 영향을 미치는지 도출.
- 강수량과 검색량 간의 상관관계를 데이터로 명확히 확인.

---

#### 5️⃣ **코드**

```python
# 날짜를 기준으로 데이터 병합 함수 정의
# 특정 기간(7월 또는 9월)의 강수량 데이터와 부침개 검색량 데이터를 날짜를 기준으로 병합
def merge_data(koreanPancakeCSVArr, weatherArr):
    # 1: 기상청 데이터 처리
    # '일시' 컬럼을 datetime 형식으로 변환하여 병합 준비
    weatherArr['날짜'] = pd.to_datetime(weatherArr['일시'])  # 문자열 -> datetime
    # 강수량 데이터에서 필요한 컬럼(날짜와 강수량)만 선택
    weatherArr = weatherArr[['날짜', '강수량(mm)']]

    # 2: 부침개 검색량 데이터 병합 준비
    merged_data = []  # 병합된 데이터를 저장할 리스트
    for csv_path in koreanPancakeCSVArr:  # 부침개 검색량 데이터를 연도별로 순회
        # CSV 파일 읽기
        search_df = pd.read_csv(csv_path)
        # 'period' 컬럼을 datetime 형식으로 변환하여 병합 준비
        search_df['날짜'] = pd.to_datetime(search_df['period'])  # 문자열 -> datetime
        # 검색량 데이터에서 필요한 컬럼(날짜와 검색량)만 선택
        search_df = search_df[['날짜', 'searchVolume']]

        # 3: 날짜를 기준으로 두 데이터를 병합
        # 강수량 데이터(weatherArr)와 검색량 데이터(search_df)를 '날짜'를 기준으로 병합
        # inner join으로 병합하여 날짜가 겹치는 데이터만 유지
        merged_df = pd.merge(weatherArr, search_df, on='날짜', how='inner')
        # 병합 결과를 리스트에 추가
        merged_data.append(merged_df)

    return merged_data  # 연도별 병합 결과를 리스트로 반환


# 7월 데이터 병합 실행
merged_data_07 = merge_data(koreanPancakeCSVArr07, weatherArr07)
# 9월 데이터 병합 실행
merged_data_09 = merge_data(koreanPancakeCSVArr09, weatherArr09)

# NaN 값을 0으로 치환하는 함수
def replace_nan_with_zero(df):
    return df.fillna(0)

# NaN 값을 0으로 치환
merged_data_07 = [replace_nan_with_zero(df) for df in merged_data_07]
merged_data_09 = [replace_nan_with_zero(df) for df in merged_data_09]

# 7월 데이터 출력
start_year = 2021  # 데이터 시작 연도
for i, df in enumerate(merged_data_07):  # 연도별 데이터 순회
    year = start_year + i  # 현재 연도 계산
    print(f"7월 데이터 ({year}년도) 일부 출력:")
    display(df.sample(n=3))  # 무작위 3개의 행 출력

# 9월 데이터 출력
for i, df in enumerate(merged_data_09):  # 연도별 데이터 순회
    year = start_year + i  # 현재 연도 계산
    print(f"9월 데이터 ({year}년도) 일부 출력:")
    display(df.sample(n=3))  # 무작위 3개의 행 출력

```


### (2) 상관관계를 통해 데이터 해석

##### 상관계수의 의미 요약

- **1.0에 가까운 양수**: 강한 양의 상관관계 (한 변수가 증가하면 다른 변수도 증가).
- **0.5 ~ 0.8**: 보통 정도의 양의 상관관계.
- **0 ~ 0.3**: 약한 양의 상관관계 또는 거의 무관함.
- **0.0**: 상관관계 없음.
- **-0.3 ~ -0.5**: 약한 음의 상관관계 (한 변수가 증가하면 다른 변수는 감소).
- **-0.5 ~ -0.8**: 보통 정도의 음의 상관관계.
- **-1.0에 가까운 음수**: 강한 음의 상관관계 (한 변수가 증가할 때 다른 변수는 감소).

---

#### 1️⃣ 목적
- 강수량과 부침개 검색량 간의 상관관계를 정량적으로 측정.
- 상관계수를 통해 강수량이 검색량 변화에 미치는 영향을 수치로 파악.
- 양적 분석을 통해 패턴 및 추세를 도출하여 유의미한 데이터 인사이트 확인.

---

#### 2️⃣ 분석 개요
- 강수량: 기상청 데이터의 날짜별 강수량(mm) 값을 기준으로 분석.
- 검색량: 네이버 검색 트렌드에서 추출된 날짜별 부침개 검색량(searchVolume).
- 상관관계: 강수량과 검색량 간의 상관계수를 연도별로 계산.
- 상관계수의 구간을 기준으로 두 변수 간의 관계를 평가.

---

#### 3️⃣ 분석 과정
- 연도별 데이터 병합: 강수량 데이터(weatherArr)와 검색량 데이터(koreanPancakeCSVArr)를 날짜를 기준으로 병합.
- 상관계수 계산: 연도별로 강수량(강수량(mm))과 검색량(searchVolume)의 상관계수를 계산.
- 상관계수 값에 따라 양의 상관관계, 음의 상관관계, 또는 관계 없음으로 분류.
- 결과 시각화: 연도별 상관계수 값을 테이블로 시각화
- 특정 연도에서 장마철, 추석 등이 영향을 미쳤는지 추가 분석.

---

#### 4️⃣ 결과 기대
- 강수량과 검색량 간의 연관성을 수치화하여 도출.
- 양의 상관관계: 강수량이 많을수록 검색량이 증가하는 경향 확인.
- 음의 상관관계: 특정 요인(코로나, 외식 감소 등)으로 인해 검색량이 감소한 경우 확인.
- 관계 없음: 강수량이 검색량에 미치는 영향이 미미한 경우 식별.

---

#### 5️⃣ 코드

```python
# 상관관계 계산 함수
# 특정 데이터(7월 또는 9월)에 대해 강수량과 검색량 간의 상관관계를 계산
def calculate_correlation(merged_data, month):
    correlations = []  # 상관관계 결과를 저장할 리스트
    for i, df in enumerate(merged_data):  # 연도별 병합 데이터를 순회
        # 강수량과 검색량 간의 상관계수를 계산
        correlation = df['강수량(mm)'].corr(df['searchVolume'])
        year = start_year + i  # 현재 연도를 계산
        correlations.append((year, correlation))  # 연도와 상관계수를 저장
    return correlations  # 상관관계 리스트 반환

# 7월 데이터 상관관계 분석
# 7월 데이터에서 연도별 강수량과 검색량 간 상관관계를 계산
correlations_07 = calculate_correlation(merged_data_07, "7월")
print("7월 데이터 강수량과 검색량의 상관관계:")
for year, corr in correlations_07:  # 연도별 상관관계를 출력
    print(f"{year}년: 상관계수 = {corr:.2f}")

# 9월 데이터 상관관계 분석
# 9월 데이터에서 연도별 강수량과 검색량 간 상관관계를 계산
correlations_09 = calculate_correlation(merged_data_09, "9월")
print("9월 데이터 강수량과 검색량의 상관관계:")
for year, corr in correlations_09:  # 연도별 상관관계를 출력
    print(f"{year}년: 상관계수 = {corr:.2f}")
```

### (3) 강수량 조건에 따른 검색량 평균 비교

---

#### 1️⃣ **목적**
- 강수량이 많은 날과 강수량이 없는 날의 검색량 평균을 비교하여 강수량이 검색량에 미치는 영향을 파악.
- 7월과 9월 데이터를 대상으로 연도별 패턴을 분석.

---

#### 2️⃣ **분석 개요**
- **강수량 많은 날**: 강수량(mm)이 0보다 큰 날을 기준으로 검색량 평균 계산.
- **강수량 없는 날**: 강수량(mm)이 0인 날을 기준으로 검색량 평균 계산.
- 두 조건의 검색량 평균을 비교하여 유의미한 차이를 도출.

---

#### 3️⃣ **분석 과정**
1. **데이터 필터링**:
   - 강수량이 0보다 큰 날의 검색량 데이터를 필터링하여 평균 계산.
   - 강수량이 0인 날의 검색량 데이터를 필터링하여 평균 계산.

2. **연도별 계산**:
   - 연도별 데이터를 순회하며 각각의 조건에 대해 검색량 평균을 계산.

3. **결과 도출**:
   - 연도별로 강수량 조건에 따른 검색량 평균 값을 도출하여 차이를 비교.

---

#### 4️⃣ **결과 기대**
- **강수량 많은 날**: 비오는 날에 부침개 검색량이 증가하는 패턴을 관찰.
- **강수량 없는 날**: 비가 없는 날에 부침개 검색량이 낮아지는 경향을 확인.
- 계절적 요인과 검색량 변화 간의 관계를 유추.

---

#### 5️⃣ **코드**

```python
# 강수량 조건에 따른 검색량 평균 비교 함수
# 특정 데이터에서 강수량이 많은 날과 강수량이 없는 날의 검색량 평균을 계산
def compare_rain_conditions(merged_data):
    results = []  # 연도별 결과를 저장할 리스트
    for i, df in enumerate(merged_data):  # 연도별 데이터 순회
        year = start_year + i  # 현재 연도 계산
        # 강수량이 있는 날의 검색량 평균 계산
        high_rain_avg = df[df['강수량(mm)'] > 0]['searchVolume'].mean()
        # 강수량이 없는 날의 검색량 평균 계산
        no_rain_avg = df[df['강수량(mm)'] == 0]['searchVolume'].mean()
        # 결과를 리스트에 저장
        results.append((year, high_rain_avg, no_rain_avg))
    return results  # 연도별 결과 반환

# 7월 데이터 분석
# 7월 데이터에서 강수량 조건에 따른 검색량 평균 계산
comparison_07 = compare_rain_conditions(merged_data_07)
print("7월 데이터 강수량 조건에 따른 검색량 평균:")
for year, high_avg, no_avg in comparison_07:  # 연도별 결과 출력
    print(f"{year}년: 강수량 많은 날 평균 = {high_avg:.2f}, 강수량 없는 날 평균 = {no_avg:.2f}")

# 9월 데이터 분석
# 9월 데이터에서 강수량 조건에 따른 검색량 평균 계산
comparison_09 = compare_rain_conditions(merged_data_09)
print("9월 데이터 강수량 조건에 따른 검색량 평균:")
for year, high_avg, no_avg in comparison_09:  # 연도별 결과 출력
    print(f"{year}년: 강수량 많은 날 평균 = {high_avg:.2f}, 강수량 없는 날 평균 = {no_avg:.2f}")
```
