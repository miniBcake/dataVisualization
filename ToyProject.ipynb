```python
# ÎÇ†ÏßúÎ•º Í∏∞Ï§ÄÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© Ìï®Ïàò Ï†ïÏùò
# ÌäπÏ†ï Í∏∞Í∞Ñ(7Ïõî ÎòêÎäî 9Ïõî)Ïùò Í∞ïÏàòÎüâ Îç∞Ïù¥ÌÑ∞ÏôÄ Î∂ÄÏπ®Í∞ú Í≤ÄÏÉâÎüâ Îç∞Ïù¥ÌÑ∞Î•º ÎÇ†ÏßúÎ•º Í∏∞Ï§ÄÏúºÎ°ú Î≥ëÌï©
def merge_data(koreanPancakeCSVArr, weatherArr):
    # 1: Í∏∞ÏÉÅÏ≤≠ Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨
    # 'ÏùºÏãú' Ïª¨ÎüºÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÌôòÌïòÏó¨ Î≥ëÌï© Ï§ÄÎπÑ
    weatherArr['ÎÇ†Ïßú'] = pd.to_datetime(weatherArr['ÏùºÏãú'])  # Î¨∏ÏûêÏó¥ -> datetime
    # Í∞ïÏàòÎüâ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÌïÑÏöîÌïú Ïª¨Îüº(ÎÇ†ÏßúÏôÄ Í∞ïÏàòÎüâ)Îßå ÏÑ†ÌÉù
    weatherArr = weatherArr[['ÎÇ†Ïßú', 'Í∞ïÏàòÎüâ(mm)']]

    # 2: Î∂ÄÏπ®Í∞ú Í≤ÄÏÉâÎüâ Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© Ï§ÄÎπÑ
    merged_data = []  # Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
    for csv_path in koreanPancakeCSVArr:  # Î∂ÄÏπ®Í∞ú Í≤ÄÏÉâÎüâ Îç∞Ïù¥ÌÑ∞Î•º Ïó∞ÎèÑÎ≥ÑÎ°ú ÏàúÌöå
        # CSV ÌååÏùº ÏùΩÍ∏∞
        search_df = pd.read_csv(csv_path)
        # 'period' Ïª¨ÎüºÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÌôòÌïòÏó¨ Î≥ëÌï© Ï§ÄÎπÑ
        search_df['ÎÇ†Ïßú'] = pd.to_datetime(search_df['period'])  # Î¨∏ÏûêÏó¥ -> datetime
        # Í≤ÄÏÉâÎüâ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÌïÑÏöîÌïú Ïª¨Îüº(ÎÇ†ÏßúÏôÄ Í≤ÄÏÉâÎüâ)Îßå ÏÑ†ÌÉù
        search_df = search_df[['ÎÇ†Ïßú', 'searchVolume']]

        # 3: ÎÇ†ÏßúÎ•º Í∏∞Ï§ÄÏúºÎ°ú Îëê Îç∞Ïù¥ÌÑ∞Î•º Î≥ëÌï©
        # Í∞ïÏàòÎüâ Îç∞Ïù¥ÌÑ∞(weatherArr)ÏôÄ Í≤ÄÏÉâÎüâ Îç∞Ïù¥ÌÑ∞(search_df)Î•º 'ÎÇ†Ïßú'Î•º Í∏∞Ï§ÄÏúºÎ°ú Î≥ëÌï©
        # inner joinÏúºÎ°ú Î≥ëÌï©ÌïòÏó¨ ÎÇ†ÏßúÍ∞Ä Í≤πÏπòÎäî Îç∞Ïù¥ÌÑ∞Îßå Ïú†ÏßÄ
        merged_df = pd.merge(weatherArr, search_df, on='ÎÇ†Ïßú', how='inner')
        # Î≥ëÌï© Í≤∞Í≥ºÎ•º Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
        merged_data.append(merged_df)

    return merged_data  # Ïó∞ÎèÑÎ≥Ñ Î≥ëÌï© Í≤∞Í≥ºÎ•º Î¶¨Ïä§Ìä∏Î°ú Î∞òÌôò


# 7Ïõî Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© Ïã§Ìñâ
merged_data_07 = merge_data(koreanPancakeCSVArr07, weatherArr07)
# 9Ïõî Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© Ïã§Ìñâ
merged_data_09 = merge_data(koreanPancakeCSVArr09, weatherArr09)

# NaN Í∞íÏùÑ 0ÏúºÎ°ú ÏπòÌôòÌïòÎäî Ìï®Ïàò
def replace_nan_with_zero(df):
    return df.fillna(0)

# NaN Í∞íÏùÑ 0ÏúºÎ°ú ÏπòÌôò
merged_data_07 = [replace_nan_with_zero(df) for df in merged_data_07]
merged_data_09 = [replace_nan_with_zero(df) for df in merged_data_09]

# 7Ïõî Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•
start_year = 2021  # Îç∞Ïù¥ÌÑ∞ ÏãúÏûë Ïó∞ÎèÑ
for i, df in enumerate(merged_data_07):  # Ïó∞ÎèÑÎ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏàúÌöå
    year = start_year + i  # ÌòÑÏû¨ Ïó∞ÎèÑ Í≥ÑÏÇ∞
    print(f"7Ïõî Îç∞Ïù¥ÌÑ∞ ({year}ÎÖÑÎèÑ) ÏùºÎ∂Ä Ï∂úÎ†•:")
    display(df.sample(n=3))  # Î¨¥ÏûëÏúÑ 3Í∞úÏùò Ìñâ Ï∂úÎ†•

# 9Ïõî Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•
for i, df in enumerate(merged_data_09):  # Ïó∞ÎèÑÎ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏàúÌöå
    year = start_year + i  # ÌòÑÏû¨ Ïó∞ÎèÑ Í≥ÑÏÇ∞
    print(f"9Ïõî Îç∞Ïù¥ÌÑ∞ ({year}ÎÖÑÎèÑ) ÏùºÎ∂Ä Ï∂úÎ†•:")
    display(df.sample(n=3))  # Î¨¥ÏûëÏúÑ 3Í∞úÏùò Ìñâ Ï∂úÎ†•

```

```python
# Í∞ïÏàòÎüâ Ï°∞Í±¥Ïóê Îî∞Î•∏ Í≤ÄÏÉâÎüâ ÌèâÍ∑† ÎπÑÍµê Ìï®Ïàò
# ÌäπÏ†ï Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í∞ïÏàòÎüâÏù¥ ÎßéÏùÄ ÎÇ†Í≥º Í∞ïÏàòÎüâÏù¥ ÏóÜÎäî ÎÇ†Ïùò Í≤ÄÏÉâÎüâ ÌèâÍ∑†ÏùÑ Í≥ÑÏÇ∞
def compare_rain_conditions(merged_data):
    results = []  # Ïó∞ÎèÑÎ≥Ñ Í≤∞Í≥ºÎ•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
    for i, df in enumerate(merged_data):  # Ïó∞ÎèÑÎ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏàúÌöå
        year = start_year + i  # ÌòÑÏû¨ Ïó∞ÎèÑ Í≥ÑÏÇ∞
        # Í∞ïÏàòÎüâÏù¥ ÏûàÎäî ÎÇ†Ïùò Í≤ÄÏÉâÎüâ ÌèâÍ∑† Í≥ÑÏÇ∞
        high_rain_avg = df[df['Í∞ïÏàòÎüâ(mm)'] > 0]['searchVolume'].mean()
        # Í∞ïÏàòÎüâÏù¥ ÏóÜÎäî ÎÇ†Ïùò Í≤ÄÏÉâÎüâ ÌèâÍ∑† Í≥ÑÏÇ∞
        no_rain_avg = df[df['Í∞ïÏàòÎüâ(mm)'] == 0]['searchVolume'].mean()
        # Í≤∞Í≥ºÎ•º Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•
        results.append((year, high_rain_avg, no_rain_avg))
    return results  # Ïó∞ÎèÑÎ≥Ñ Í≤∞Í≥º Î∞òÌôò

# 7Ïõî Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
# 7Ïõî Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í∞ïÏàòÎüâ Ï°∞Í±¥Ïóê Îî∞Î•∏ Í≤ÄÏÉâÎüâ ÌèâÍ∑† Í≥ÑÏÇ∞
comparison_07 = compare_rain_conditions(merged_data_07)
print("7Ïõî Îç∞Ïù¥ÌÑ∞ Í∞ïÏàòÎüâ Ï°∞Í±¥Ïóê Îî∞Î•∏ Í≤ÄÏÉâÎüâ ÌèâÍ∑†:")
for year, high_avg, no_avg in comparison_07:  # Ïó∞ÎèÑÎ≥Ñ Í≤∞Í≥º Ï∂úÎ†•
    print(f"{year}ÎÖÑ: Í∞ïÏàòÎüâ ÎßéÏùÄ ÎÇ† ÌèâÍ∑† = {high_avg:.2f}, Í∞ïÏàòÎüâ ÏóÜÎäî ÎÇ† ÌèâÍ∑† = {no_avg:.2f}")

# 9Ïõî Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
# 9Ïõî Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í∞ïÏàòÎüâ Ï°∞Í±¥Ïóê Îî∞Î•∏ Í≤ÄÏÉâÎüâ ÌèâÍ∑† Í≥ÑÏÇ∞
comparison_09 = compare_rain_conditions(merged_data_09)
print("9Ïõî Îç∞Ïù¥ÌÑ∞ Í∞ïÏàòÎüâ Ï°∞Í±¥Ïóê Îî∞Î•∏ Í≤ÄÏÉâÎüâ ÌèâÍ∑†:")
for year, high_avg, no_avg in comparison_09:  # Ïó∞ÎèÑÎ≥Ñ Í≤∞Í≥º Ï∂úÎ†•
    print(f"{year}ÎÖÑ: Í∞ïÏàòÎüâ ÎßéÏùÄ ÎÇ† ÌèâÍ∑† = {high_avg:.2f}, Í∞ïÏàòÎüâ ÏóÜÎäî ÎÇ† ÌèâÍ∑† = {no_avg:.2f}")
```
